**Выделение признаков item_features (вакансии)**

```python
# Группировка данных по дате и вакансии, а затем по cookie_id, и выбор только записей с event_type = 1
grouped_data = data[data['event_type'] == 1].groupby(['event_date', 'vacancy_id_', 'cookie_id'])
grouped_data
```

```python
grouped_data.nunique()['event_type']
```

```python
# Подсчет количества уникальных cookie_id, выполнивших событие event_type = 1
conversion = grouped_data.nunique()['event_type'].groupby(['event_date', 'vacancy_id_']).sum()
conversion.reset_index()
```

```python
# Подсчет общего количества уникальных cookie_id для каждой вакансии
unique_cookies = data.groupby(['event_date', 'vacancy_id_'])['cookie_id'].nunique()
unique_cookies.reset_index()
```

```python
daily_conversion = pd.merge(unique_cookies.reset_index(), conversion.reset_index(), \
                            on=['event_date', 'vacancy_id_'], how='outer')
daily_conversion
```

```python
daily_conversion['event_type'].fillna(0, inplace=True)
```

```python
# конверсия ежедневная
daily_conversion['day_conv'] = daily_conversion['event_type'] / daily_conversion['cookie_id']
daily_conversion
```

```python
daily_conversion[(daily_conversion['day_conv'] > 0) &(daily_conversion['day_conv'] < 1)]
```

```python
# Первое взаимодействие и крайнее взаимодействие с вакансиями.
group_vac = daily_conversion.groupby('vacancy_id_').agg({'event_date': ['last', 'first']}).reset_index()
group_vac
```

```python
group_vac.iloc[:, 1:3] = group_vac.iloc[:, 1:3].astype('datetime64')
```

```python
group_vac.info()
```

```python
data[data['vacancy_id_'] == 260164]
```

```python
# Расчет разницы в днях между взаимодействиями по вакансиям
group_vac[('diff_days', '')] = (group_vac[('event_date', 'last')] - group_vac[('event_date', 'first')]).dt.days
```

```python
group_vac
```

```python
group_vac[group_vac[('diff_days', '')] == 1]
```

```python
group_vac[('diff_days', '')] = group_vac[('diff_days', '')].replace({0: 1})
```

```python
# Суммирование конверсий.
conv_day = daily_conversion.groupby('vacancy_id_')['day_conv'].sum().reset_index()
conv_day
```

```python
conv_day_ = pd.merge(conv_day, group_vac, on='vacancy_id_', how='outer')
```

```python
conv_day_
```

```python
# Нахождение средней дневной конверсии
conv_day_['avg_day_conv'] = round(conv_day_['day_conv'] / conv_day_[('diff_days', '')], 2)
```

```python
conv_day_[['vacancy_id_', 'avg_day_conv']]
```

```python
conv_day_['avg_day_conv'].value_counts()
```

```python
pd.cut(conv_day_['avg_day_conv'], 5).value_counts()
```

```python
# Разбивка на 5 категорий среднедневных конверсий.
conv_day_['cat_conv'] = pd.cut(conv_day_['avg_day_conv'], 5, labels = ['c1', 'c2', 'c3', 'c4', 'c5'])
```

```python
conv_day_
```

```python
conv_day_[conv_day_['avg_day_conv'] == 1]
```

```python
# Запись признака item_feature категории конверсии в csv
conv_day_[['vacancy_id_', 'cat_conv']].to_csv(os.path.join('item_feature_conv.csv'), index=False)
```

```python
# Запись признака item_feature конверсии в csv
conv_day_[['vacancy_id_', 'avg_day_conv']].to_csv(os.path.join('item_feature_avg_conv.csv'), index=False)
```

```python
from datetime import datetime
end_date = datetime(2022, 10, 1)
end_date-datetime(2022, 9, 29)
```

```python
# Расчет признака item_feature "коэффициент затухания".
conv_day_['time_decay'] = conv_day_[('event_date', 'last')].apply(lambda x: np.exp(-(end_date - x).days))
```

```python
# Группировка по cookie, вакансиям, max взаимодействие.
data_copy_train = data_copy.groupby(['cookie_id', 'vacancy_id_']).agg({'event_type': 'max'}).reset_index()
data_copy_train
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data_copy_train, test_size=0.3, random_state=42)
train
```

```python
test
```

```python
# Выделяем уникальные cookie в разделенном тесте посредством train_test_split, у которых более 4 вакансий с взаимодействиями.
nv_test = test.groupby('cookie_id').agg({'vacancy_id_':'nunique'}).reset_index()
set_test_cookie = nv_test[nv_test['vacancy_id_'] >= 5]
test_cookie = set(set_test_cookie['cookie_id'])
test_cookie
```

```python
nv_test[nv_test['vacancy_id_'] < 5]
```

```python
# Выбираем строки с cookie из cookies_set
df_test_selected = test[test['cookie_id'].isin(test_cookie)]
df_test_selected
```

```python
df_test_selected['cookie_id'].nunique()
```

```python
# Выбираем строки с cookie, которых нет в cookies_set
df_test_not_selected = test[~test['cookie_id'].isin(test_cookie)]
df_test_not_selected
```

```python
# Объединяем трейн и не вошедшие в тест куки с вакансиями
df_train_selected = pd.concat([train, df_test_not_selected])
df_train_selected
```

```python
# Запись в csv train
df_train_selected.to_csv(os.path.join('train_data.csv'), index=False)
```

```python
# Запись в csv test
df_test_selected.to_csv(os.path.join('test_data.csv'), index=False)
```
